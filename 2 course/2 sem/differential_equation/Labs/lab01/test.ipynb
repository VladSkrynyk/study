```python
import numpy as np
from matplotlib import pyplot as plt


y = np.arange(0.01, 2.8, 0.01) 
plt.fill_between(2 - y*y - 1 / (2*y), y, color = 'none', hatch="o", edgecolor="g")
plt.xlim([-6, 6])
plt.plot(2 - y*y - 1 / (2*y), y, color="magenta")
```
